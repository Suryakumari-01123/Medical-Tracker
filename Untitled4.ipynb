```python
import json
import pandas as pd
import re
from collections import Counter

# Load your JSON dataset
with open('mental_health_data.json', 'r') as f:
    raw_data = json.load(f)

print(f"Raw dataset size: {len(raw_data)}")

def clean_mental_health_data_fast(data):
    """Fast cleaning for mental health dataset"""
    cleaned_data = []
    valid_conditions = {'Anxiety', 'Depression', 'Stress', 'Normal', 'Panic', 'Bipolar', 'PTSD', 'OCD'}
    seen_inputs = set()  # Use set for O(1) duplicate checking

    # Condition mapping for faster lookup
    condition_mapping = {
        'anxious': 'Anxiety', 'anxiousness': 'Anxiety',
        'depressed': 'Depression', 'depressive': 'Depression',
        'stressed': 'Stress', 'stressful': 'Stress',
        'panic attack': 'Panic', 'panic disorder': 'Panic',
        'ptsd': 'PTSD', 'post-traumatic stress': 'PTSD',
        'obsessive-compulsive': 'OCD', 'ocd': 'OCD',
        'bipolar disorder': 'Bipolar', 'manic': 'Bipolar',
        'healthy': 'Normal', 'fine': 'Normal', 'good': 'Normal'
    }

    for item in data:
        # Quick validation - skip malformed items
        if not isinstance(item, dict) or 'input' not in item or 'output' not in item:
            continue

        # Handle input
        input_value = item.get('input')
        if not input_value or (isinstance(input_value, float) and pd.isna(input_value)):
            continue

        # Convert and clean input
        input_text = str(input_value).strip()
        if len(input_text) < 5:
            continue

        # Simple regex cleaning (faster than multiple re.sub calls)
        input_text = re.sub(r'[^\w\s\.\,\!\?\'\-]+', ' ', input_text)
        input_text = re.sub(r'\s+', ' ', input_text).strip()

        # Fast duplicate check with set
        input_lower = input_text.lower()
        if input_lower in seen_inputs:
            continue
        seen_inputs.add(input_lower)

        # Handle output
        output_value = item.get('output')
        if not output_value or (isinstance(output_value, float) and pd.isna(output_value)):
            continue

        output_condition = str(output_value).strip()
        if not output_condition or output_condition.lower() in ('nan', 'none', ''):
            continue

        # Fast condition mapping
        output_lower = output_condition.lower()
        if output_lower in condition_mapping:
            output_condition = condition_mapping[output_lower]
        else:
            output_condition = output_condition.title()

        # Skip invalid conditions
        if output_condition not in valid_conditions:
            continue

        cleaned_data.append({
            'input': input_text,
            'output': output_condition,
            'instruction': str(item.get('instruction', 'Classify the mental health condition based on the statement.')).strip()
        })

    return cleaned_data

# Clean the data (should be fast now)
import time
start_time = time.time()
cleaned_data = clean_mental_health_data_fast(raw_data)
end_time = time.time()

print(f"Cleaned dataset size: {len(cleaned_data)}")
print(f"Cleaning took: {end_time - start_time:.2f} seconds")

# Show distribution
condition_counts = Counter(item['output'] for item in cleaned_data)
print("\nCondition distribution:")
for condition, count in sorted(condition_counts.items()):
    print(f"{condition}: {count}")

# Convert to conversational format
def format_for_chat(sample):
    return {
        "messages": [
            {
                "role": "user",
                "content": f"{sample['instruction']}\n\nStatement: {sample['input']}"
            },
            {
                "role": "assistant",
                "content": f"Based on the statement provided, this appears to indicate: **{sample['output']}**\n\nPlease remember that this is an AI assessment and should not replace professional mental health consultation. If you're experiencing distress, consider speaking with a qualified mental health professional."
            }
        ]
    }

# Apply formatting
formatted_data = [format_for_chat(item) for item in cleaned_data]

# Create dataset and split
from datasets import Dataset
dataset = Dataset.from_list(formatted_data)
dataset = dataset.train_test_split(test_size=0.1, shuffle=True, seed=42)

print(f"Training samples: {len(dataset['train'])}")
print(f"Validation samples: {len(dataset['test'])}")
print(f"Total processing time: {time.time() - start_time:.2f} seconds")
```

```python
from unsloth import FastLanguageModel
from transformers import TrainingArguments, Trainer, DataCollatorForLanguageModeling
from datasets import Dataset
import torch

# Step 1: Load Model and Apply LoRA
max_seq_length = 2048
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/gemma-3-270m-it",
    max_seq_length = max_seq_length,
    dtype = None,
    load_in_4bit = True,
)

model = FastLanguageModel.get_peft_model(
    model,
    r = 16,
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    random_state = 3407,
)

# Setup tokenizer
tokenizer.pad_token = tokenizer.eos_token

# Step 2: Prepare Your Dataset
def format_conversation(sample):
    """Format sample into training text"""
    messages = [
        {"role": "user", "content": f"{sample['instruction']}\n\nStatement: {sample['input']}"},
        {"role": "assistant", "content": f"Based on the statement provided, this appears to indicate: **{sample['output']}**\n\nPlease remember that this is an AI assessment and should not replace professional mental health consultation. If you're experiencing distress, consider speaking with a qualified mental health professional."}
    ]

    # Create conversation text
    conversation = ""
    for message in messages:
        if message["role"] == "user":
            conversation += f"<start_of_turn>user\n{message['content']}<end_of_turn>\n"
        else:
            conversation += f"<start_of_turn>model\n{message['content']}<end_of_turn>\n"

    return {"text": conversation}

# Apply to your cleaned_data (replace with your actual 64,734 samples)
formatted_data = [format_conversation(sample) for sample in cleaned_data]
dataset = Dataset.from_list(formatted_data)
dataset = dataset.train_test_split(test_size=0.1, shuffle=True, seed=42)

print(f"Training samples: {len(dataset['train'])}")
print(f"Validation samples: {len(dataset['test'])}")

# Step 3: Critical Fix - Manual Tokenization
def tokenize_function(examples):
    """Tokenize with proper padding and truncation"""
    result = tokenizer(
        examples["text"],
        truncation=True,
        padding="max_length",
        max_length=max_seq_length,
        return_tensors=None,
    )
    # For causal language modeling, labels = input_ids
    result["labels"] = result["input_ids"].copy()
    return result

# Apply tokenization
train_dataset = dataset['train'].map(tokenize_function, batched=True, remove_columns=["text"])
eval_dataset = dataset['test'].map(tokenize_function, batched=True, remove_columns=["text"])

print("✅ Dataset tokenized successfully")

# Step 4: Use Standard Trainer (Not SFTTrainer)
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

training_args = TrainingArguments(
    output_dir="gemma-270m-therapist",
    per_device_train_batch_size=1,
    gradient_accumulation_steps=8,
    warmup_steps=5,
    max_steps=2000,  # Adjust for your 64K dataset
    learning_rate=2e-4,
    fp16=True,
    logging_steps=50,
    save_steps=500,
    save_total_limit=3,
    dataloader_num_workers=0,
    remove_unused_columns=False,
    report_to=[],
    seed=3407,
    optim="adamw_8bit",
    weight_decay=0.01,
    lr_scheduler_type="linear",
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    data_collator=data_collator,
)

# Step 5: Start Training
print("🚀 Starting training...")
trainer.train()

# Step 6: Save Model
model.save_pretrained("gemma-270m-therapist-final")
tokenizer.save_pretrained("gemma-270m-therapist-final")

print("✅ Training completed and model saved!")
```
